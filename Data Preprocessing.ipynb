## Importing Libraries for visualisation and preprocessing

```python
import numpy as np 
import pandas as pd 
from sklearn import preprocessing
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

## Reading the data

```python
data=pd.read_csv('data.csv')
data.head()
```

```python
data.drop(columns=['id','Unnamed: 32'], inplace=True)
data.head()
```

## Checking for NaN values

```python
data.isna().sum()
```

## Checking the distribution of classes

```python
sns.countplot(data.diagnosis,label="Count")
```

```python
y=data['diagnosis']
data.drop(columns=['diagnosis'], inplace=True)
```

## Normalizing the data

```python
mm_scaler = preprocessing.MinMaxScaler()
data[data.columns] = mm_scaler.fit_transform(data[data.columns])
data.head()
```

## Plotting the correlation matrix

```python
corr=data.corr().round(3)
f, ax = plt.subplots(figsize=(20, 20))
sns.heatmap(corr, annot=True)
```

Some columns are highly correlated with each other, eg. radius_mean column has a correlation of 1 and 0.99 with perimeter_mean and area_mean columns. This is probably because the three columns essentially contain the same information. It is better to pick only one of these three columns.
Also, there is multicollinearity present between all the "mean" and "worst" columns. This is probably because to calculated the mean, we need to consider the worst value of the respective column. Thus, we do not need these worst values as well.

```python
data = data[data.columns.drop(list(data.filter(regex='worst')))]
data = data[data.columns.drop(list(data.filter(regex='perimeter')))]
data = data[data.columns.drop(list(data.filter(regex='area')))]
```

## New correlation matrix

```python
corr=data.corr().round(3)
f, ax = plt.subplots(figsize=(20, 20))
sns.heatmap(corr, annot=True)
```
